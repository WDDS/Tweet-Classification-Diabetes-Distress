```python
import pandas as pd
nutr_path = "/home/adrian/PhD/Data/SocioeconomicEnvironmentalFactors/Nutrition__Physical_Activity__and_Obesity_-_Women__Infant__and_Child.csv"

data_nutr = pd.read_csv(nutr_path, sep=";")

data_nutr = data_nutr[data_nutr.YearEnd == 2014]
data_nutr = data_nutr[data_nutr.StratificationCategory1 == "Total"]
#data_nutr = data_nutr[data_nutr.DataValueType == "Crude Prevalence"]


print(data_nutr.columns)
print()

data_nutr = data_nutr[['YearEnd', 'LocationAbbr', 'LocationDesc', 'Datasource',
       'Class', 'Topic', 'Question', 'Data_Value', 'Data_Value_Alt', 
       'Sample_Size']]
print(data_nutr.Question.value_counts())
data_nutr.head()
```

```python
var1 = data_nutr[data_nutr["Question"] == "Percent of WIC children aged 2 to 4 years who have an overweight classification"][["LocationDesc", "Data_Value"]]
var1.columns = ["LocationDesc", "Obesity-Weight__2014__Percent-of-WIC-children-aged-2-to-4-years-who-have-an-overweight-classification"]
var2 = data_nutr[data_nutr["Question"] == "Percent of WIC children aged 2 to 4 years who have obesity"][["LocationDesc", "Data_Value"]]
var2.columns = ["LocationDesc", "Obesity-Weight__2014__Percent-of-WIC-children-aged-2-to-4-years-who-have-obesity"]
var3 = data_nutr[data_nutr["Question"] == "Percent of WIC children aged 3-23 months old who have a high weight-for-length"][["LocationDesc", "Data_Value"]]
var3.columns = ["LocationDesc", "Obesity-Weight__2014__Percent-of-WIC-children-aged-3-23-months-old-who-have-a-high-weight-for-length"]

var2.head()

joined_obesity = var1.merge(var2,on='LocationDesc').merge(var3,on='LocationDesc')
joined_obesity.index = joined_obesity.LocationDesc.values
joined_obesity.drop("LocationDesc", axis=1, inplace=True)
joined_obesity.head()
```

```python
# https://catalog.data.gov/dataset/air-quality-measures-on-the-national-environmental-health-tracking-network
air_path = "/home/adrian/PhD/Data/SocioeconomicEnvironmentalFactors/Air_Quality_Measures_on_the_National_Environmental_Health_Tracking_Network.csv"

data_air = pd.read_csv(air_path, sep=",")

data_air = data_air[data_air.ReportYear == 2013] # 2011
#data_air = data_air[data_air.MeasureName == "Number of days with maximum 8-hour average ozone concentration over the National Ambient Air Quality Standard"]
#data_air = data_air[data_air.StateName == "Washington"]


print(data_air.columns)
print()
print(data_air.ReportYear.value_counts())
print()
print(data_air.MeasureType.value_counts())
print()
print(data_air.UnitName.value_counts())

print()
print(data_air.MeasureName.value_counts())
data_air.head(50)
```

```python
#airQuality = pd.DataFrame()

var1 = data_air[data_air.MeasureName == "Number of person-days with maximum 8-hour average ozone concentration over the National Ambient Air Quality Standard"]
#print(var1.Value.value_counts())
meanVar1 = var1.groupby(by="StateName")["Value"].mean()
meanVar1_pd = pd.DataFrame(meanVar1.values, index=meanVar1.index, columns=["Number_of_person-days_with_maximum_8-hour_average_ozone_concentration_over_the_National_Ambient_Air_Quality_Standard__MeanPerStateName"])

var2 = data_air[data_air.MeasureName == "Number of days with maximum 8-hour average ozone concentration over the National Ambient Air Quality Standard"]
#print(var2.Value.value_counts())
meanVar2 = var2.groupby(by="StateName")["Value"].mean()
meanVar2_pd = pd.DataFrame(meanVar2.values, index=meanVar2.index, columns=["Number_of_days_with_maximum_8-hour_average_ozone_concentration_over_the_National_Ambient_Air_Quality_Standard__MeanPerStateName"])

var3 = data_air[data_air.MeasureName == "Percent of days with PM2.5 levels over the National Ambient Air Quality Standard (NAAQS)"]
#print(var3.Value.value_counts())
meanVar3 = var3.groupby(by="StateName")["Value"].mean()
meanVar3_pd = pd.DataFrame(meanVar3.values, index=meanVar3.index, columns=["Percent_of_days_with_PM2.5_levels_over_the_National_Ambient_Air_Quality_Standard_(NAAQS)"])

var4 = data_air[data_air.MeasureName == "Person-days with PM2.5 over the National Ambient Air Quality Standard"]
#print(var3.Value.value_counts())
meanVar4 = var4.groupby(by="StateName")["Value"].mean()
meanVar4_pd = pd.DataFrame(meanVar4.values, index=meanVar4.index, columns=["Person-days_with_PM2.5_over_the_National_Ambient_Air_Quality_Standard"])

var5 = data_air[data_air.MeasureName == "Annual average ambient concentrations of PM2.5 in micrograms per cubic meter (based on seasonal averages and daily measurement)"]
#print(var3.Value.value_counts())
meanVar5 = var5.groupby(by="StateName")["Value"].mean()
meanVar5_pd = pd.DataFrame(meanVar5.values, index=meanVar5.index, columns=["Annual_average_ambient_concentrations_of_PM2.5_in_micrograms_per_cubic_meter_(based_on_seasonal_averages_and_daily_measurement)"])

joined_air = pd.concat([meanVar1_pd, meanVar2_pd, meanVar2_pd, meanVar2_pd, meanVar2_pd], axis = 1)
joined_air.head()
```
